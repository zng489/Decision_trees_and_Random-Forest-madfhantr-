```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn import neighbors, datasets
```

```python
df = pd.read_csv('madfhantr.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Gender'] = df['Gender'].replace(to_replace='Male',value='1')
print(df['Gender'])

df['Gender'] = df['Gender'].replace(to_replace='Female',value='0')
print(df['Gender'])
```

```python
df['Married'] = df['Married'].replace(to_replace='Yes',value='1')
print(df['Married'])

df['Married'] = df['Married'].replace(to_replace='No',value='0')
print(df['Married'])
```

```python
df['Self_Employed'] = df['Self_Employed'].replace(to_replace='No',value='0')
df['Self_Employed'] = df['Self_Employed'].replace(to_replace='Yes',value='1')
```

```python
df['Education'] = df['Education'].replace(to_replace='Graduate',value='1')
df['Education'] = df['Education'].replace(to_replace='Not Graduate',value='0')
```

```python
df['Dependents'] = df['Dependents'].replace(to_replace='3+',value='3')
```

# Splitting the data-set X and y

```python
X = df.drop(columns=['Loan_Status','Loan_ID','Property_Area'])
X
```

```python
y = df.Loan_Status
y
```

# Building the model X and y

# Decision Trees

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(max_depth=3)
clf.fit(X_train,y_train)
```

```python
predictions = clf.predict(X_test)
predictions
```

```python
type(predictions)
```

```python
print(np.count_nonzero(predictions == 'Y'))
```

```python
print(np.count_nonzero(predictions == 'N'))
```

```python
clf.score(X_test,y_test)
```

```python
# Acuracia
print(99/(31+99)*100)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))


# matriz de confusão
# (TP + TN)/total

(99+13)/(13+99+1+99)*100
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)
```
